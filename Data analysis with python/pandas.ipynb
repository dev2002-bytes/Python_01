```python
import pandas as pd
```

```python
##PANDAS SERIES IS A ONE-DIMENSIONAL LABELED ARRAY CAPABLE OF HOLDING ANY DATA TYPE (INTEGERS, STRINGS, FLOATS, PYTHON OBJECTS, ETC.). IT IS SIMILAR TO A NUMPY ARRAY BUT WITH ADDITIONAL FUNCTIONALITY AND FLEXIBILITY. A SERIES CAN BE CREATED FROM A LIST, DICTIONARY, OR ANOTHER ARRAY-LIKE STRUCTURE. IT HAS AN INDEX THAT LABELS EACH ELEMENT IN THE SERIES, ALLOWING FOR EASY ACCESS AND MANIPULATION OF DATA. PANDAS SERIES ARE USEFUL FOR HANDLING AND ANALYZING DATA IN A VARIETY OF APPLICATIONS, INCLUDING TIME SERIES ANALYSIS, DATA CLEANING, AND STATISTICAL MODELING.
```

```python
import pandas as pd
data=[10,20,30,40,50]
series=pd.Series(data)
print(series)
```

```python
## create series from dictionary
data={'a':10,'b':20,'c':30}     ## a=key and 10=value that is assigned to key a
series=pd.Series(data)
print(series)
```

```python
data=[10,20,30]
index=['a','b','c']
series=pd.Series(data,index=index)
print(series)
```

```python
##dataframe is a two-dimensional labeled data structure with columns of potentially different types. It is similar to a spreadsheet or SQL table, and it is one of the most commonly used data structures in pandas. A DataFrame can be created from various data sources, such as lists, dictionaries, or even other DataFrames. It provides powerful tools for data manipulation, including filtering, grouping, and aggregation. With DataFrames, you can easily handle and analyze large datasets, making it an essential tool for data scientists and analysts.
```

```python
## create dataframe from dictionary
data={'name':['Alice','Bob','Charlie'],'age':[25,30,35],'city':['New York','Los Angeles','Chicago']}
df=pd.DataFrame(data)
print(df)
```

```python
import numpy as np
np.array(df)
```

```python
## create a dataframe from a list of dictionaries

data=[{'name':'Alice','age':25,'city':'New York'},{'name':'Bob','age':30,'city':'Los Angeles'},{'name':'Charlie','age':35,'city':'Chicago'}]
df=pd.DataFrame(data)
print(df)


df.iloc[0]
df.loc[0][2]
```

```python
df.loc[0]['city']
```

```python
df.at[0,'city']  ## access a single value for a row/column label pair using the at method, which is faster than loc for accessing a single value.
```

```python
df.at[2,'name']
```

```python
## accessing a specified elements using iat

df.iat[0,2]  ## access a single value for a row/column pair by integer position using the iat method, which is faster than iloc for accessing a single value.
```
